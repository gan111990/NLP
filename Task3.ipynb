```python
class WeatherForecaster():

    def __init__(self, country = 'Novosibirsk, RU'):
        self.country = country
        self.create_conversations()
        !wget -nc https://raw.githubusercontent.com/chucky995/DL_NLP/master/tf-idf-faq.json
        self.bot = train_model("tf-idf-faq.json", download = True)
        
    def ask(self, questions):
        faq_answers = self.bot(questions)
        answers = []

        for i, score in enumerate(np.array(faq_answers[1])):
            if score.max()>0.9:
                answers.append(faq_answers[0][i][0])
            else:
                answers.append("Sorry! Don't know the answer to that!")
        return answers
      
      
    def create_conversations(self):
        days = {0:"today", 1:"tomorrow", 2:"day after tomorrow", 3:"in three days", 4:"in four days"}
        owm = OWM('81eafa57483f61466c11276914ba8728')
        thf = owm.three_hours_forecast(self.country)

        start_date = thf.when_starts('date')
        questions = []

        for i in days:
            weather_i = thf.get_weather_at(thf.when_starts('date') + timedelta(days = i))

            questions += [["What is the weather like " + days[i] + " ?", weather_i.get_detailed_status()],
                             ["How is the weather " + days[i] + " ?", weather_i.get_detailed_status()],
                             ["What is the temperature " + days[i] + " ?", weather_i.get_temperature(unit='celsius').get('temp')],
                             ["How much degrees outside " + days[i] + " ?",  weather_i.get_temperature(unit='celsius').get('temp')],
                             ["How much wind " + days[i] + " ?", weather_i.get_wind()],
                             ["What are the wind details for" + days[i] + " ?", weather_i.get_wind()],
                             ["What is the precipitation level" + days[i] + " ?",  weather_i.get_rain()],
                             ["How much rain " + days[i] + " ?", weather_i.get_rain()]]

        df = pd.DataFrame(questions, columns=['Question', 'Answer'])
        print(df.head())
        #df.to_csv('faq_en.csv', index = 'False', sep=',', line_terminator='\n')
```

```python
bot = WeatherForecaster()
print(bot.ask(["What is the weather like today?"]))
```

```python
def weather_at_day(day, carac):  
  owm = pyowm.OWM('81eafa57483f61466c11276914ba8728')
  thf = owm.three_hours_forecast('Novosibirsk, RU')
  weather = thf.get_weather_at(thf.when_starts('date') + timedelta(days = day))  
  
  if (carac == 'weather' or carac =='погода'):
    return weather.get_detailed_status()
  if (carac == 'temperarure' or carac == 'degrees' or carac == 'температура' or carac =='градусов'):
    return weather.get_temperature(unit='celsius').get('temp')
  if (carac == 'wind' or 'ветра'):
    return weather.get_wind()
  if (carac == 'precipitation' or carac == 'rain' or carac == 'осадков'):
    return weather.get_rain()

class ActionGetToday(Action):
    def name(self):
      return 'action_today'

    def run(self, dispatcher, tracker, domain):
      t = tracker.get_slot('carac')
      
      if (t == None):
        message = "Sorry! Don't know the answer to that!"
      else:
        message = weather_at_day(0, t)

      dispatcher.utter_message(str(message))
      
class ActionGetTomorrow(Action):
    def name(self):
      return 'action_tomorrow'

    def run(self, dispatcher, tracker, domain):
      t = tracker.get_slot('carac')
      
      if (t == None):
        message = "Sorry! Don't know the answer to that!"
      else:
        message = weather_at_day(1, t)

      dispatcher.utter_message(str(message))
      
class ActionGetTwo(Action):
    def name(self):
      return 'action_two'

    def run(self, dispatcher, tracker, domain):
      t = tracker.get_slot('carac')
      
      if (t == None):
        message = "Sorry! Don't know the answer to that!"
      else:
        message = weather_at_day(2, t)
      
      dispatcher.utter_message(str(message))

class ActionGetThree(Action):
    def name(self):
      return 'action_three'

    def run(self, dispatcher, tracker, domain):
      t = tracker.get_slot('carac')
      
      if (t == None):
        message = "Sorry! Don't know the answer to that!"
      else:
        message = weather_at_day(3, t)
      
      dispatcher.utter_message(str(message))
      
class ActionGetFour(Action):
    def name(self):
      return 'action_four'

    def run(self, dispatcher, tracker, domain):
      t = tracker.get_slot('carac')
      
      if (t == None):
        message = "Sorry! Don't know the answer to that!"
      else:
        message = weather_at_day(4, t)

      dispatcher.utter_message(str(message))
```

```python
training_data = load_data('nlu-data-en.md')
trainer = Trainer(config.load('config_en.yml'))
interpreter = trainer.train(training_data)
model_directory = trainer.persist('model/nlu', fixed_model_name="current")
```

```python
fallback = FallbackPolicy(fallback_action_name="utter_unclear",
                          core_threshold=0.8,
                          nlu_threshold=0.6)
agent = Agent('domain_en.yml', policies=[MemoizationPolicy(), KerasPolicy(), fallback])
training_data = agent.load_data('stories.md')
agent.train(training_data, epochs = 100)
agent.persist('model/dialogue')
```

```python
interpreter = NaturalLanguageInterpreter.create(model_directory)
agent = Agent.load('model/dialogue', interpreter=interpreter)
```

```python
responses = agent.handle_message('How is the weather tomorrow?')
for response in responses:
    print(response["text"])
```
